```python
import numpy as np
import pandas as pd

cat_df = pd.read_csv("cat_breeds_dirty.csv", sep=";")
cat_df.head()
```

```python
#Information
cat_df.info()
```

```python
#Stats
cat_df.describe()
```

# Rename Columns First

```python
#Proper column naming
cat_df.columns = cat_df.columns.str.strip().str.lower().str.replace(" ", "_")

#Drop latitude and longitude columns, and age in years
cat_df.drop(columns=["latitude", "longitude"], inplace=True)
cat_df.head()
```

# Clean Rows

```python
cat_df.isnull().sum()
```

### Breed Cleaning

```python
#Current unique breeds
unique_breeds = cat_df["breed"].unique()
print(f"Current Unique Breeds:")
print(unique_breeds)

#Replace misspelled breed names
cat_df["breed"] = cat_df["breed"].replace(["Ankora", "angora", "Angorra"], "Angora")
cat_df["breed"] = cat_df["breed"].replace(["Maine loon", "maine coon", "Maine coon"], "Maine Coon")
cat_df["breed"] = cat_df["breed"].replace(["ragdoll", "rack doll", "wrack doll"], "Ragdoll")

#Replacing nan
breed_mode = cat_df["breed"].mode(dropna=True)[0]
cat_df["breed"] = cat_df["breed"].fillna(breed_mode)

#New unique breeds
unique_breeds = cat_df["breed"].unique()
print(f"New Unique Breeds:")
print(unique_breeds)
```

### Age Cleaning

```python
#Check null values in age in years and months
print("Null values in age_in_years:", cat_df["age_in_years"].isnull().sum())
print("Null values in age_in_months:", cat_df["age_in_months"].isnull().sum())
both_null_count = (cat_df["age_in_years"].isnull() & cat_df["age_in_months"].isnull()).sum()
print("Null values in both age_in_years and age_in_months:", both_null_count)
print("="*100)

#Add mode-based age (in years) for rows with both null age in years and null age in months
age_years_mode = cat_df["age_in_years"].mode(dropna=True)[0]
age_months_mode = cat_df["age_in_months"].mode(dropna=True)[0]
cat_df.loc[cat_df["age_in_years"].isnull() & cat_df["age_in_months"].isnull(), "age_in_years"] = age_years_mode

#Clean Age in years and months
print("Null values in age_in_years:", cat_df["age_in_years"].isnull().sum())
both_null_count = (cat_df["age_in_years"].isnull() & cat_df["age_in_months"].isnull()).sum()
print("Null values in both age_in_years and age_in_months:", both_null_count)
print("="*100)

#Fill null months by multiplying years by 12
cat_df.loc[cat_df["age_in_months"].isnull(), "age_in_months"] = cat_df["age_in_years"] * 12

#Fill null years by dividing months by 12
cat_df.loc[cat_df["age_in_years"].isnull(), "age_in_years"] = cat_df["age_in_months"] / 12

#Check null values again
print("Null values in age_in_years:", cat_df["age_in_years"].isnull().sum())
print("Null values in age_in_months:", cat_df["age_in_months"].isnull().sum())
```

### Gender Cleaning

```python
unique_genders =cat_df["gender"].unique()
print(f"Unique Genders:")
print(unique_genders)

#Replace missing genders
cat_df.loc[cat_df["gender"].isnull(), "gender"] = np.random.choice(["male", "female"])

#Final check for null values
unique_genders =cat_df["gender"].unique()
print(f"Unique Genders:")
print(unique_genders)
```

### Cleaning Neutered or Spayed Column

```python
unique_neutered_or_spayed = cat_df["neutered_or_spayed"].unique()
print(f"Unique Neutered or Spayed:")
print(unique_neutered_or_spayed)

#Replace missing genders
mode_neutered_or_spayed = cat_df["neutered_or_spayed"].mode(dropna=True)[0]
cat_df.loc[cat_df["neutered_or_spayed"].isnull(), "neutered_or_spayed"] = mode_neutered_or_spayed

#Final check for null values
unique_neutered_or_spayed = cat_df["neutered_or_spayed"].unique()
print(f"Unique Neutered or Spayed:")
print(unique_neutered_or_spayed)
```

### Cleaning Body Length and Weight Columns

```python
#Check null values in body length and weight
print("Null values in body_length:", cat_df["body_length"].isnull().sum())
print("Null values in weight:", cat_df["weight"].isnull().sum())


#Fill missing body lengths with mean
mean_body_length = cat_df["body_length"].mean()
cat_df["body_length"] = cat_df["body_length"].fillna(mean_body_length)

#Fill missing weights with mean
mean_weight = cat_df["weight"].mean()
cat_df["weight"] = cat_df["weight"].fillna(mean_weight)

#Final check for null values
print("="*100)
print("Null values in body_length:", cat_df["body_length"].isnull().sum())
print("Null values in weight:", cat_df["weight"].isnull().sum())
```

### Cleaning Dominant Fur Color, Fur Pattern, and Eye Color

```python
#Unique Dominant Colors
unique_dominant_colors = cat_df["fur_colour_dominant"].unique()
print(f"Unique Dominant Colors:")
print(unique_dominant_colors)

#Unique Fur Patterns
unique_fur_patterns = cat_df["fur_pattern"].unique()
print(f"Unique Fur Patterns:")
print(unique_fur_patterns)

#Unique Eye Colors
unique_eye_colors = cat_df["eye_colour"].unique()
print(f"Unique Eye Colors:")
print(unique_eye_colors)
```

```python
#Replace unknown and unusual values
cat_df["fur_colour_dominant"] = cat_df["fur_colour_dominant"].replace(["what does it mean dominant?"], np.nan)
cat_df["fur_colour_dominant"] = cat_df["fur_colour_dominant"].replace(["red/cream"], "cream")
cat_df["fur_colour_dominant"] = cat_df["fur_colour_dominant"].replace(["brown/chocolate"], "brown")

cat_df["eye_colour"] = cat_df["eye_colour"].replace(["cute", "I dont know. Its pretty!"], np.nan)

# Find the mode, ignoring nan
fur_colour_dominant_mode = cat_df['fur_colour_dominant'].mode(dropna=True)[0]
fur_pattern_mode = cat_df['fur_pattern'].mode(dropna=True)[0]
eye_colour_mode = cat_df['eye_colour'].mode(dropna=True)[0]

# Replace all np.nan, 'UNKNOWN', and 'ERROR' with the mode
cat_df['fur_colour_dominant'] = cat_df['fur_colour_dominant'].fillna(fur_colour_dominant_mode)
cat_df['fur_pattern'] = cat_df['fur_pattern'].fillna(fur_pattern_mode)
cat_df['eye_colour'] = cat_df['eye_colour'].fillna(eye_colour_mode)

#Then check again

#Unique Dominant Colors
unique_dominant_colors = cat_df["fur_colour_dominant"].unique()
print(f"Unique Dominant Colors:")
print(unique_dominant_colors)

#Unique Fur Patterns
unique_fur_patterns = cat_df["fur_pattern"].unique()
print(f"Unique Fur Patterns:")
print(unique_fur_patterns)

#Unique Eye Colors
unique_eye_colors = cat_df["eye_colour"].unique()
print(f"Unique Eye Colors:")
print(unique_eye_colors)

print("="*100)
print("New Null values in fur_colour_dominant:", cat_df["fur_colour_dominant"].isnull().sum())
print("New Null values in fur_pattern:", cat_df["fur_pattern"].isnull().sum())
print("New Null values in eye_colour:", cat_df["eye_colour"].isnull().sum())
```

### Cleaning Allowed Outdoor

```python
#Current unique values in the 'allowed_outdoor' column:
print("Unique values in 'allowed_outdoor' column:")
print(cat_df["allowed_outdoor"].unique())

#Replace not TRUE and not FALSE values with np.nan
cat_df["allowed_outdoor"] = cat_df["allowed_outdoor"].replace(["I never allow my kitty outside!!!!!", "I dont allow her outside. I'm a responsible owner"], np.nan)
allowed_outdoor_mode = cat_df["allowed_outdoor"].mode(dropna=True)[0]
cat_df["allowed_outdoor"] = cat_df["allowed_outdoor"].fillna(allowed_outdoor_mode)

#New values
print("="*100)
print("New unique values in 'allowed_outdoor':")
print(cat_df["allowed_outdoor"].unique())
print("New Null values in 'allowed_outdoor':", cat_df["allowed_outdoor"].isnull().sum())
```

### Clean Preferred Food

```python
print("Current Unique Preferred Foods:")
print(cat_df["preferred_food"].unique())
print("Null values in preferred_food:", cat_df["preferred_food"].isnull().sum())

#Replace values
cat_df["preferred_food"] = cat_df["preferred_food"].replace(["a lot of food"], np.nan)
preferred_food_mode = cat_df["preferred_food"].mode(dropna=True)[0]
cat_df["preferred_food"] = cat_df["preferred_food"].fillna(preferred_food_mode)

#New values
print("="*100)
print("New unique values in 'preferred_food':")
print(cat_df["preferred_food"].unique())
print("Null values in preferred_food:", cat_df["preferred_food"].isnull().sum())
```

### Clean Playtime

```python
#Current unique owner play times
print(f"Null Playtimes: {cat_df['owner_play_time_minutes'].isnull().sum()}")

#Replacing nan with median
play_time_median = cat_df["owner_play_time_minutes"].median(skipna=True)
cat_df["owner_play_time_minutes"] = cat_df["owner_play_time_minutes"].fillna(play_time_median)

#New unique owner play times
print(f"New Null Playtimes: {cat_df['owner_play_time_minutes'].isnull().sum()}")
```

### Cleaning Sleep Time

```python
#Current unique owner play times
print(f"Null Sleep Times: {cat_df['sleep_time_hours'].isnull().sum()}")

#Replacing nan with median
play_time_median = cat_df["sleep_time_hours"].median(skipna=True)
cat_df["sleep_time_hours"] = cat_df["sleep_time_hours"].fillna(play_time_median)

#New unique owner play times
print(f"New Null Sleep Times: {cat_df['sleep_time_hours'].isnull().sum()}")
```

### Cleaning Country Column

```python
#Current unique countries
unique_countries = cat_df["country"].unique()
print(f"Current Unique Countries:")
print(unique_countries)
print("Current Null Countries:", cat_df["country"].isnull().sum())

#Replace misspelled country names
cat_df["country"] = cat_df["country"].replace(["Vive la France!", "La France!!!!", "france"], "France")

#Replace unknown countries
cat_df["country"] = cat_df["country"].replace(["with me", "where I live", "my country"], np.nan)


#Replacing nan with mode
country_mode = cat_df["country"].mode(dropna=True)[0]
cat_df["country"] = cat_df["country"].fillna(country_mode)

#New unique countries
print("="*100)
unique_countries = cat_df["country"].unique()
print(f"Current Unique Countries:")
print(unique_countries)
print("Current Null Countries:", cat_df["country"].isnull().sum())
```
